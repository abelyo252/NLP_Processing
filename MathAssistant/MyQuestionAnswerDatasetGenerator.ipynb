```python
def generate_dataset(num_samples, num_specific_words):
    dataset = []
    unrecognized_answer = "I'm sorry, your query does not appear to be a recognized term that I'm familiar with."
    word_list = nltk.corpus.words.words()

    specific_words = random.sample(word_list, num_specific_words)

    for _ in range(num_samples):
        random_word = random.choice(word_list + specific_words)
        question = "Who is " + random_word + "?"
        questions.append(question)
        answers.append(unrecognized_answer)
```

```python
# Create question and answer directories if they don't exist
os.makedirs("question", exist_ok=True)
os.makedirs("answer", exist_ok=True)
```

```python

# Save each question and answer to separate text files
for i in tqdm(range(len(questions)), desc="Saving files", unit="file"):
    question_filename = f"question/question_{i + 1}.txt"
    answer_filename = f"answer/answer_{i + 1}.txt"

    # Check if the file already exists
    if os.path.exists(question_filename):
        # Generate a unique filename by appending random alphanumeric characters
        random_suffix = ''.join(random.choices(string.ascii_lowercase + string.digits, k=6))
        question_filename = f"question/question_{i + 1}_{random_suffix}.txt"

    if os.path.exists(answer_filename):
        random_suffix = ''.join(random.choices(string.ascii_lowercase + string.digits, k=6))
        answer_filename = f"answer/answer_{i + 1}_{random_suffix}.txt"

    with open(question_filename, "w") as question_file:
        question_file.write(questions[i])
    with open(answer_filename, "w") as answer_file:
        answer_file.write(answers[i])

    tqdm.write(f"Saved question {i + 1} to {question_filename}")
    tqdm.write(f"Saved answer {i + 1} to {answer_filename}")
```
